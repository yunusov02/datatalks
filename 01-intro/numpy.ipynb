## Randomly Generated Arrays

```python
np.random.seed(2) # seed the random number generator
np.random.rand(5, 4)
```

```python
np.random.randn(5, 4)
```

```python
np.random.randint(1, 10, (5, 4)) # low, high, size
```

```python
a = np.arange(1, 5)
a
```

```python
a + 1
```

```python
a >= 3
```

```python
a = np.random.randint(2, 10, (1, 5))
b = np.random.randint(2, 10, (1, 5))


print(a)
print(b)
print(a >= b)
print(a[a >= b])

```

```python
a = np.random.randint(2, 50, (100,))

print(a)
print("Sum:", a.sum())
print("Mean:", a.mean())
print("Max:", a.max())
print("Min:", a.min())
```
